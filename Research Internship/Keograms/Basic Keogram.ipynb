```python
# Importing Libraries
from ampere_utils.plot import polar_plot, configure_polar_plot
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import xarray as xr
import datetime as dt
```

```python
# Loading Ampere Data:
filepath = r"C:\Users\Furio\Documents\University\Master's\Internship\Data" # Change Appropriately
Ampere_file_location = filepath + r"\AMPERE\grd\2015"
Ampere_file = Ampere_file_location + r"\ampere.20150101.0000.86400.120.north.grd.ncdf"
Ampere_data = xr.load_dataset(Ampere_file)
Ampere_data
```

```python
# Loading SuperDARN Data:
SuperDARN_file = r"C:\Users\Furio\Documents\University\Master's\Internship\Data\SuperDARN\2015\01\sd_vorticity_20150101_n.nc"
SuperDARN_data = xr.load_dataset(SuperDARN_file)
SuperDARN_data
```

```python
# Removing invalid values
SuperDARN = xr.where(SuperDARN_data['vorticity'] >= 9990, np.nan, SuperDARN_data['vorticity'])
Ampere = xr.where(Ampere_data['jPar'] >= 9990, np.nan, Ampere_data['jPar'])
```

```python
Ampere = Ampere.values.reshape(720, 24, 50)
```

```python
Ampere.shape
```

```python
j = Ampere
x = SuperDARN

time = np.array([dt.datetime(2015, 1, 1) + dt.timedelta(seconds = int(np.round(((t*3600))))) for t in Ampere_data.time.data])
indices = np.linspace(0, (len(time) - 1), num=4, dtype=int)
tick_positions = time[indices]   
tick_labels = [time[i] for i in indices] 


J_colorbar_limit = 0.4
j_norm = mcolors.TwoSlopeNorm(vmin=-J_colorbar_limit, vcenter=0, vmax=J_colorbar_limit)

x_colorbar_limit = 0.2 
x_norm = mcolors.TwoSlopeNorm(vmin=-x_colorbar_limit, vcenter=0, vmax=x_colorbar_limit)

fig,ax = plt.subplots(2,1, figsize=(15,8))
plot_1 = ax[0].pcolormesh(time, np.arange(0,50,1), j[:,6,:].T, cmap = "RdBu_r", norm = j_norm)
ax[0].set_title("Ampere")
ax[0].set_xticks(tick_positions, tick_labels)
ax[0].set_xlabel("Time")
ax[0].set_ylabel("Colat")
plt.colorbar(plot_1, ax = ax[0], extend = "both")

plot_2 = ax[1].pcolormesh(time[0:-2],np.arange(0,40,1),x[:,6,:].T, cmap = "RdBu_r", norm = x_norm)
ax[1].set_title("SuperDARN")
ax[1].set_xticks(tick_positions, tick_labels)
ax[1].set_xlabel("Time")
ax[1].set_ylabel("Colat")
plt.colorbar(plot_2, ax = ax[1], extend = "both")

plt.tight_layout()
plt.show()
```
